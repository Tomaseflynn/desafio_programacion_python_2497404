```python
#!pip install requests
import requests

if __name__== "__main__":
  url= 'https://api.mercadolibre.com/sites/MLA/search?q="Paragolpes%20Tector%20Iveco"&attributes=id,title,condition,price,sold_quantity,available_quantity'
  respuesta= requests.get(url)

  if respuesta.status_code==200:
    contenido= respuesta.content
    print(contenido)
```

```python
import json

# String completo
string_completo = contenido


# Cargar el string como objeto JSON
json_objeto = json.loads(string_completo)

# Recorrer los valores del objeto JSON y reemplazar la cadena "Cordoba" por "Córdoba"
def reemplazar_cordoba(objeto):
    if isinstance(objeto, dict):
        for clave, valor in objeto.items():
            if isinstance(valor, str):
                objeto[clave] = valor.replace("\xc3\xb3", "Córdoba")
            elif isinstance(valor, (dict, list)):
                reemplazar_cordoba(valor)
    elif isinstance(objeto, list):
        for i in range(len(objeto)):
            if isinstance(objeto[i], str):
                objeto[i] = objeto[i].replace("\xc3\xb3", "Córdoba")
            elif isinstance(objeto[i], (dict, list)):
                reemplazar_cordoba(objeto[i])

# Llamar a la función para reemplazar la cadena
reemplazar_cordoba(json_objeto)

# Convertir el objeto JSON modificado de nuevo a string
string_modificado = json.dumps(json_objeto)

# Imprimir el string modificado
print(string_modificado)
```
